```python
import math
import matplotlib.pyplot as plt
import networkx as nx
n = 8846 
pvals=[]
avgDegree = []
NGbyN = []
for i in range(0,51):
    ptemp = (i/10)*(1/8845)
    pvals.append(ptemp)
    avgDegree.append(i/10)
pvals.append(0.001027325)
avgDegree.append(9.087720)
for p in pvals:
    G = nx.binomial_graph(n, p)
    # identify largest connected component
    Gcc = sorted(nx.connected_components(G), key=len, reverse=True)
    G0 = G.subgraph(Gcc[0])
    NGbyN.append(len(G0.nodes())/8846)
```

```python
print(avgDegree)
print(NGbyN)
plt.plot(avgDegree, NGbyN)
```

```python
ebet = nx.edge_betweenness_centrality(G)
import operator
sorted_ebet = sorted(ebet.items(), key=operator.itemgetter(1), reverse=True)
sorted_ebet[0:5]
```

```python
K2 = G.copy()

def remove_top_ebet(K2):
    ebet = nx.edge_betweenness_centrality(K2)
    sorted_ebet = sorted(ebet.items(), key=operator.itemgetter(1), reverse=True)
    edge_to_remove = sorted_ebet[0]
    K2.remove_edge(*edge_to_remove[0])

    return K2
```

```python
pos = nx.fruchterman_reingold_layout(G);
```

```python
import matplotlib.pyplot as plt
for i in range(5):
    print("Communities after", i+1, "th iteration:")
    K2=remove_top_ebet(K2)
    nx.draw(K2, pos)
    nx.draw_networkx_labels(K2, pos)
    plt.show()
```
